### Problem Statement and Objective: 
- Build a machine learning model to classify iris flowers into one of the three species (setosa, versicolor, or virginica) based on the given features (sepal length, sepal width, petal length, and petal width). The goal is to create a model that can accurately predict the species of an iris flower based on its measurements.
- The objective of this project is to develop a machine learning model capable of accurately classifying iris flowers into their respective species—setosa, versicolor, or virginica—based on four measured features: sepal length, sepal width, petal length, and petal width.


### Iris Dataset: 
- The Iris dataset consists of measurements of various characteristics of iris flowers. It includes three species of iris flowers: setosa, versicolor, and virginica. For each species, four features are measured, resulting in a total of 150 observations (50 samples from each species). The four features are:

    - Sepal Length: Length of the iris flower's sepals (the outermost whorl of a flower, usually green and protective).
    - Sepal Width: Width of the iris flower's sepals.
    - Petal Length: Length of the iris flower's petals (the innermost whorl of a flower, often colorful and attractive).
    - Petal Width: Width of the iris flower's petals.
    
- The dataset is available in many machine learning libraries, including scikit-learn in Python.

### Step 2: Data Cleaning
- Missing values handling
- Duplicate entries check
- Outliers handling

```python
import pandas as pd

# Check for missing values
missing_values = iris_df.isnull().sum()
print("Missing values:\n", missing_values)



# Check for duplicate entries
duplicate_entries = iris_df.duplicated().sum()
print("\nNumber of duplicate entries:", duplicate_entries)



# Check for outliers - lets plot boxplot to get better view
# columns to include in the box plot
columns_to_plot = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create a box plot for each specified column
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df[columns_to_plot])
plt.title('Box Plot for Iris Dataset Features')
plt.show()
```

```python
# Calculate IQR for sepal width
Q1 = iris_df['SepalWidthCm'].quantile(0.25)
Q3 = iris_df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outliers
outliers = iris_df[(iris_df['SepalWidthCm'] < lower_bound) | (iris_df['SepalWidthCm'] > upper_bound)]
iris_df = iris_df[~((iris_df['SepalWidthCm'] < lower_bound) | (iris_df['SepalWidthCm'] > upper_bound))]
```

```python
# Create a box plot for each specified column
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df[columns_to_plot])
plt.title('Box Plot for Iris Dataset Features')
plt.show()
```

```python
iris_df.shape
```

#### Insights:
- The dataset is free from any instances of null values.
- No duplicate entries are present within the dataset.
- Outliers have been identified in the "sepal width cm" variable of the dataset.
- Outliers in the "sepal width cm" variable have been successfully eliminated through the application of the Interquartile Range (IQR) method.
- Box plots were used to visualize variable distributions before and after applying the IQR method, demonstrating successful removal of outliers in the "sepal width cm" variable.
- After handling with outliers we have shape - (146, 6) of dataset.

### Step 4: Train - Test Split of dataset

```python
# Putting feature variable to X
X = iris_df.drop(['Species', 'Id'],axis=1)

# Putting response variable to y
y = iris_df['Species']
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
X_train.shape, X_test.shape
```

### Step 5: Modeling:  Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# Importing required packages for visualization
from IPython.display import Image  
from six import StringIO  
from sklearn.tree import export_graphviz
import pydotplus, graphviz
```

```python
# plotting decision tree 

dot_data = StringIO()  

export_graphviz(dt, out_file=dot_data, filled=True, rounded=True,
                feature_names=X.columns, 
                class_names=['Iris-virginica', 'Iris-versicolor', 'Iris-setosa'])



graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

### Evaluating model performance

```python
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print("Train - Accuracy score:", accuracy_score(y_train, y_train_pred))
print ("Train - Confusion Matrix:", confusion_matrix(y_train, y_train_pred))
```

```python
print("Test - Accuracy score:", accuracy_score(y_test, y_test_pred))
print("Test - Confusion Matrix:", confusion_matrix(y_test, y_test_pred))
```

### Step 6: Decision Tree Hyper-parameter tuning

```python
dth = DecisionTreeClassifier(random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Create the parameter grid based on the results of random search 
params = {
    'max_depth': [2, 3, 5, 10],
    'min_samples_leaf': [5, 10, 20, 50],
    'criterion': ["gini", "entropy"]
}
```

```python
# Instantiate the grid search model
grid_search = GridSearchCV(estimator=dth, 
                           param_grid=params, 
                           cv=4, n_jobs=-1, verbose=1, scoring = "accuracy")
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
dt_best = grid_search.best_estimator_
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, dt_best.predict(X_test)))
```

```python
# plotting decision tree 

dot_data = StringIO()  

export_graphviz(dt_best, out_file=dot_data, filled=True, rounded=True,
                feature_names=X.columns, 
                class_names=['Iris-virginica', 'Iris-versicolor', 'Iris-setosa'])



graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
